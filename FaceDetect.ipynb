```python
from pathlib import Path
import pandas as pd
import glob

data_dir = Path("/kaggle/input/fer2013")

if not data_dir.exists():
    raise FileNotFoundError(f"{data_dir} does not exist")

print("Top-level files and folders:")
for p in sorted(data_dir.iterdir()):
    print("-", p.name)

csv_files = list(data_dir.glob("**/*.csv"))
if csv_files:
    print("\nFound CSV files:")
    for f in csv_files:
        print("-", f)
    main_csv = next((f for f in csv_files if "fer" in f.name.lower()), csv_files[0])
    print(f"\nLoading {main_csv} into a DataFrame (showing first 5 rows):")
    df = pd.read_csv(main_csv)
    print(df.head())
else:
    img_paths = []
    for ext in ("**/*.jpg", "**/*.jpeg", "**/*.png"):
        img_paths.extend(list(data_dir.glob(ext)))
    print(f"\nFound {len(img_paths)} image files.")
    for p in img_paths[:20]:
        print("-", p)
```
